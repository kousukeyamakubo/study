```python
import matplotlib.pyplot as plt

# ---  ---
labels = ['Without Tracking', 'With Tracking']
values = [5.8747, 1.3045]

# ---  ---
# 
plt.figure(figsize=(10, 8))

# 
bars = plt.bar(labels, values, color=['#4c72b0', '#55a868'], width=0.6)

# ---  ---
# 1. Y
plt.ylabel('RMSE of Velocity [m/s]', fontsize=35) # 40

plt.tick_params(axis='x', labelsize=24, pad=20) 
plt.tick_params(axis='y', labelsize=20)

# 2. XY
plt.tick_params(axis='x', labelsize=35) # 40
plt.tick_params(axis='y', labelsize=35)

# 3. 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 7)

# 4. 5.8744
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.05,
             f'{height:.4f}',
             ha='center', va='bottom', fontsize=40) # 40

# 
#plt.tight_layout()
plt.savefig('2GHz_vel_comparison.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# ---  ---
labels = ['Without Tracking', 'With Tracking']
means = [0.0535, 0.0251]       # 
errors = [0.0017, 0.0028]      #  ()

# ---  ---
plt.figure(figsize=(10, 8))

# 
# yerr=errors capsize=10 
bars = plt.bar(labels, means, yerr=errors, capsize=10, 
               color=['#4c72b0', '#55a868'], width=0.6)

# Y
# 0.40 0.45 
plt.ylim(0, 0.45)

# ---  ---
plt.ylabel('RMSE of Range [m]', fontsize=35) # Y

# 
plt.tick_params(axis='x', labelsize=35, pad=20) # pad=20
plt.tick_params(axis='y', labelsize=35)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# --- 2 ---
for bar, err in zip(bars, errors):
    height = bar.get_height()
    # (height + err + )
    plt.text(bar.get_x() + bar.get_width()/2., height + err + 0.005,
             f'{height:.4f}\n({err:.4f})',  # \n 
             ha='center', va='bottom', fontsize=40)

plt.tight_layout()

# PNG
plt.savefig('2GHz_range_comparison.png', dpi=300)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
t = np.linspace(0, 1, 1000)
# sin(t^2)
signal = np.sin(2 * np.pi * (5 * t + 15 * t**2))

# 
plt.figure(figsize=(6, 2)) # 
plt.plot(t, signal, color='black', linewidth=2)
plt.axis('off') # 
plt.tight_layout()

# 
plt.savefig("chirp_wave.png", transparent=True, dpi=300)
plt.show()
```
