```python
import numpy as np
```

```python
# 2GHz
Tc = 0.509*1e-9
T_symbol = 1.115 * 1e-6
offset = 4000
N_sample = int(np.floor(T_symbol/Tc))
rx_sample = offset + N_sample
T_trans = rx_sample * Tc

l_speed = 299792458
f_carrier = 28 * 1e+9
lam = l_speed / f_carrier

BW = 1.966080e+9  
SNR = 10**(-3/10)
shanonn_capacity = BW * np.log2(1 + SNR)

R= 1e+9
alpha = R/shanonn_capacity
print("占有率：",alpha)
T_sensing = (1-alpha)*5e-2

N_trans = T_sensing/T_trans
print("チャープ数：",N_trans)
delta_v = lam/(2 * N_trans *T_trans)
print("速度分解能：",delta_v)
print("---占有率を個人で設定---")
alpha = 0.9
print("alpha：",alpha)
print("このときのスループット：",shanonn_capacity*alpha)
print("このときのチャープ数：",(1-alpha)*5e-2/T_trans)
```
